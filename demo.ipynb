## Explore

```python
import dask.dataframe
import dask.array
from xarray.core.indexes import PandasIndex
```

```python
no_idx_ds = ds.drop_indexes('time').drop('time')
```

```python
no_idx_ds
```

```python
no_idx_ds['dim_var'] = ('time', dask.array.from_array(ds.time.values))
```

```python
no_idx_ds
```

```python
no_idx_ds.set_coords('dim_var').sel(time=slice("2023-06-01T00:00:00.000000000", "2023-06-01T00:00:00.099999744"))
```

```python
final_ds = no_idx_ds.set_coords('dim_var').set_xindex("dim_var", PandasIndex)
```

```python
dv = final_ds.dim_var
```

```python
dv.sortby('time')
```

```python
final_ds.sortby('time').sel(dim_var=slice("2023-06-01T00:00:00.000000000", "2023-06-01T00:00:00.099999744"))
```
